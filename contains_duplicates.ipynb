````python
from typing import List
class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        hashnet = set()
        for n in nums:
            if n in hashnet:
                return True 
            hashnet.add(n)
        return False
````

````python

a = [1, 2, 3, 4, 5, 1]
b = [1, 3, 2, 4, 5, 6]


sol = Solution ()

print(sol.hasDuplicate(a))
print(sol.hasDuplicate(b))
````

````python
# Brute force method 
from typing import List
class Solution:
    def hasDuplicates(self, nums: List[int]) -> bool:
        for i in range(len(nums)):
            for j in range (i+1, len(nums)):
                if nums[i] == nums [j]:
                    return True
        return False
````

#### in here we are having a nested loop which might take the time complexity of o(n2) but we are not using any extra memory so the space complexity still would be o(1)

````python
# Hash set method

class Solution:
    def hasDupliate1(self, nums: List[int]) -> bool:
        newset = set()
        for num in nums:
            if num in newset:
                return True
            newset.add(num)
        return False
````

#### in here we are just looking up the number whther it is in the hashset so the time complexity would be o(1) cause we are only checking once, memory would be o(n) cause we store everything in the set
