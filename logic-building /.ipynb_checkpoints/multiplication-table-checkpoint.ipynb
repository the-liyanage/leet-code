```python
def main():
    number = int(input("Enter the number:"))
    mul_table(number)

def mul_table(number, i=1):
    if (i == 11):
        return
    print(f"{number} * {i} = {number * i}")
    mul_table(number, i+1)

if __name__ == "__main__":
    main()
```
