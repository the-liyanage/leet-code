```python

def main():
    number = int(input("Enter the number: "))
    sum_all(number)

def sum_all(number):
    total = 0
    for num in range(1, number+1):
        total += num
    print(total)

if __name__ == "__main__":
    main()
```

```python
def main():
    number = int(input("Enter the number:"))
    print(sum_all(number))
    
def sum_all(number):
    if number == 1:
        return number
    return number + sum_all(number -1)

if __name__ == "__main__":
    main()
```
