```python
def main():
    number = int(input("Enter your number: "))
    print(sum_square(number))

def sum_square(number):
    total = 0
    for num in range(1, number+1):
        total += num * num
    return total

if __name__ == "__main__":
    main()
```

```python
def main():
    number = int(input("Enter your number:"))
    print(sum_square(number))
    
def sum_square(number):
    if number == 1:
        return number
    return number * number + sum_square(number-1)

if __name__ == "__main__":
    main()
```
