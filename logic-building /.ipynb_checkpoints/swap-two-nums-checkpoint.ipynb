```python
a = int(input("Value for a:"))
b = int(input("Value for b:"))

temp = a
a = b
b = temp

print(f"Value for a is {a}, Value for b is {b}")
```

```python
# Without using a third variable

def main():
    a = int(input("Value for a:"))
    b = int(input("Value for b:"))
    print(swap_values(a,b))

def swap_values(a,b):
    a = a+b
    b = a-b
    a = a-b
    return f"New Value for a is {a} & New Value for b is {b}"

if __name__ == "__main__":
    main()
```

```python
# Using Tuple unpacking

def main():
    a = int(input("Value of a:"))
    b = int(input("Value of b:"))
    print(swap_values(a,b))

def swap_values(a,b):
    a, b = b, a
    return f"New Value for a is {a} & New Value for b is {b}"

if __name__ == "__main__":
    main()
```
